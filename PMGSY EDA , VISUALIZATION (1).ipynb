### INSIGHTS


### 1 Which state has sanctioned the highest number of road works?

```python
group1 = trisha.groupby("STATE_NAME").agg({"NO_OF_ROAD_WORK_SANCTIONED": "sum"})
group2 = group1.sort_values(by="NO_OF_ROAD_WORK_SANCTIONED", ascending=False)
group2
```

```python
# Plot
plt.figure(figsize=(10,6))
group2.plot(kind="bar", legend=False, color="green", edgecolor="black", width=0.8)
plt.title('Number of Sanctioned Road Works by State')
plt.xlabel('State')
plt.ylabel('Number of Sanctioned Road Works')
plt.xticks(rotation=45, ha="right", fontsize = 7) 
plt.tight_layout() 
plt.show()
```

Inference:
The bar chart illustrates the Number of Road Works Sanctioned by State. From the chart:
â€¢
Uttar Pradesh has the highest number of road works sanctioned, with around 20,000 projects.
â€¢
Madhya Pradesh, Bihar, and Rajasthan follow closely behind, each with between 15,000 and 19,000 sanctioned road works.
â€¢
States like Haryana, Nagaland, Mizoram, Ladakh, Andaman and Nicobar Islands, Goa, and Puducherry have significantly fewer sanctioned road works, below 1,000.
â€¢
There is a wide disparity in the number of road works sanctioned between states, highlighting different levels of infrastructure focus or project prioritization.
Result:
The state with the highest number of road works sanctioned is Uttar Pradesh, followed by Madhya Pradesh, Bihar, and Rajasthan.

### 2 Which state has completed the highest percentage of sanctioned road works?

```python
# Calculate completion percentage
trisha['COMPLETION_PERCENTAGE'] = (trisha['NO_OF_ROAD_WORKS_COMPLETED'] / trisha['NO_OF_ROAD_WORK_SANCTIONED']) * 100
trisha
```

```python
# Group by state and calculate the mean completion percentage
state_completion_percentage = trisha.groupby('STATE_NAME')['COMPLETION_PERCENTAGE'].mean()
state_completion_percentage 
```

```python
#Find the state with the highest percentage
max_completion_state = state_completion_percentage.idxmax()
max_completion_value = state_completion_percentage.max()
print(f"State with highest completion percentage: {max_completion_state} ({max_completion_value:.2f}%)")
```

```python
# Plot
plt.figure(figsize=(10,6))
state_completion_percentage.plot(kind='bar', color='green')
plt.title('Completion Percentage of Road Works by State')
plt.xlabel('State')
plt.ylabel('Completion Percentage (%)')
plt.xticks(rotation=90)
plt.show()
```

Inference:
Based on the chart, the states are ranked by the percentage of road works completed relative to those sanctioned. The chart shows that the majority of states are close to 100% completion. However, one state stands out with a completion rate significantly lower than the others, indicating a lag in progress.
Result:
From the chart, West Bengal appears to have the highest percentage of completed road works, reaching close to 100%, while several other states also demonstrate near-complete progress, indicated by full or nearly full bars in the graph.

### 3. How does the number of sanctioned road works correlate with the number of road works completed?

```python
# Plot a scatter plot to visualize the correlation
plt.figure(figsize=(8,6))
plt.scatter(trisha['NO_OF_ROAD_WORK_SANCTIONED'], trisha['NO_OF_ROAD_WORKS_COMPLETED'], color='green')
plt.title('Correlation Between Sanctioned and Completed Road Works')
plt.xlabel('Sanctioned Road Works')
plt.ylabel('Completed Road Works')
plt.grid(True)
plt.show()

```

```python
 #Calculate correlation
correlation = trisha['NO_OF_ROAD_WORK_SANCTIONED'].corr(trisha['NO_OF_ROAD_WORKS_COMPLETED'])
print(f"Correlation between sanctioned and completed road works: {correlation}")
```

Inference:
â€¢
The number of sanctioned road works directly influences the number of completed road works, as seen by the high correlation.
â€¢
A perfect correlation (1.00) implies that the number of sanctioned and completed works are very tightly aligned, with little or no discrepancy between what is sanctioned and what gets completed.
This visualization effectively communicates that road projects are being completed almost in line with the number of projects sanctioned.
Result:
The relationship between sanctioned and completed road works appears to be linear, with most points lying close to the line of best fit.

### 4. Which state has the highest expenditure on road projects, and how does it compare with the number of roads completed?

```python
q1 = trisha.groupby("STATE_NAME").agg({"EXPENDITURE_OCCURED_LAKHS": "sum","NO_OF_ROAD_WORKS_COMPLETED": "sum"})
q2 = q1.sort_values(by=["EXPENDITURE_OCCURED_LAKHS","NO_OF_ROAD_WORKS_COMPLETED"], ascending=False)
q2
```

Inference:
â€¢
West Bengal has the highest expenditure on road projects, indicating a significant allocation of resources towards road infrastructure.
â€¢
With 732 completed roads, the state's expenditure is aligned with a substantial number of road completions, suggesting that the higher expenditure is correlated with a high volume of completed projects.
Result:
â€¢
State with the highest expenditure: West Bengal, with an expenditure of 2004.87157 Lakhs.
â€¢
Number of roads completed in West Bengal: 732

### 5. What is the relationship between the length of road works sanctioned and the length completed across different states?

```python
# Group by state and sum sanctioned and completed lengths
q3 = trisha.groupby("STATE_NAME").agg({"LENGTH_OF_ROAD_WORK_SANCTIONED_KM": "sum","LENGTH_OF_ROAD_WORK_COMPLETED_KM": "sum"})
q4 = q3.sort_values(by=["LENGTH_OF_ROAD_WORK_SANCTIONED_KM","LENGTH_OF_ROAD_WORK_COMPLETED_KM"], ascending=False)
q4
```

Inference:

West Bengal has the highest expenditure (2004.87 Lakhs) but a moderate number of completed roads (732) compared to other states. This suggests that, while a significant amount of funds are allocated, the number of completed projects is not proportionally high.
â€¢
Uttar Pradesh and Bihar show a more efficient use of funds, with a higher number of completed roads relative to their expenditure.
â€¢
Gujarat and Haryana have high expenditures but relatively fewer completed roads, indicating potential inefficiencies or delays in project execution.
Result:
â€¢
West Bengal leads in expenditure but does not have the highest number of completed roads, raising concerns about efficiency.
â€¢
Uttar Pradesh and Bihar appear more cost-effective, completing more roads with less expenditure.
â€¢
States like Gujarat and Haryana need to assess project management practices to improve road completion relative to the funds spent.

### 10. Which state has the lowest ratio of expenditure to the cost of sanctioned works?

```python
# Calculate the ratio of expenditure to sanctioned cost
trisha['EXPENDITURE_TO_COST_RATIO'] = trisha['EXPENDITURE_OCCURED_LAKHS'] / trisha['COST_OF_WORKS_SANCTIONED_LAKHS']

# Group by state and calculate the average ratio
state_ratio = trisha.groupby('STATE_NAME')['EXPENDITURE_TO_COST_RATIO'].mean()

# Find the state with the lowest ratio
min_ratio_state = state_ratio.idxmin()
min_ratio_value = state_ratio.min()
print(f"State with the lowest expenditure to sanctioned cost ratio: {min_ratio_state} ({min_ratio_value:.2f})")


```

```python

# Plot
plt.figure(figsize=(10,6))
state_ratio.plot(kind='bar', color='lightblue')
plt.title('Expenditure to Sanctioned Cost Ratio by State')
plt.xlabel('State')
plt.ylabel('Expenditure to Cost Ratio')
plt.xticks(rotation=90)
plt.show()
```

Inference:
â€¢
The data is grouped by State Name and the expenditure that occurred for the road works and the cost that have been sanctioned for each state summed up.
â€¢
The ratio column has been inserted to the grouped data and the ratio of each has been calculated.
â€¢
The plotting of bar graph for the same that helps to identify the State which has the maximum and minimum ratio
Result:
â€¢
Goa and Andaman and Nicobar Islands have the lowest ratios of expenditure to sanctioned cost (around 0.4 and 0.5 respectively). This indicates that a smaller proportion of the sanctioned cost has been spent in these states, suggesting possible under-expenditure or delays in project implementation.
â€¢
Karnataka, Assam, and Odisha are at the top of the chart, with a ratio close to or equal to 1. This means that nearly all of the sanctioned cost has been utilized.
